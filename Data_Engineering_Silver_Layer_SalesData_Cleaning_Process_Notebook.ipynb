# **Data Transformation**

**Calendar**

```python
df_cal = spark.read.format("csv").option("header","true").load("abfss://27c585a4-80a1-47b7-879c-9a8e709ed796@onelake.dfs.fabric.microsoft.com/e5d2dce1-920c-4c8b-bb30-641e70bdd3d0/Files/rawAdventureWorks_Calendar/AdventureWorks_Calendar.csv")
# df now is a Spark DataFrame containing CSV data from "abfss://27c585a4-80a1-47b7-879c-9a8e709ed796@onelake.dfs.fabric.microsoft.com/e5d2dce1-920c-4c8b-bb30-641e70bdd3d0/Files/rawAdventureWorks_Calendar/AdventureWorks_Calendar.csv".
display(df_cal)
```

```python
df_cal= df_cal.withColumn("Date", to_date("date", "M/d/yyyy"))


display(df_cal)
```

```python
df_cal= df_cal.withColumn("Month",month(col("Date")))\
              .withColumn("Year",year(col("Date")))

display(df_cal)
```

```python
df_cal.write.format("delta")\
            .mode("append")\
            .option("path","Files/Calendar")\
            .saveAsTable("Calendar")
```

**Customers**

```python
df_cust= spark.read.format("csv")\
                   .option("header",True)\
                   .option("inferschema",True)\
                   .load("abfss://27c585a4-80a1-47b7-879c-9a8e709ed796@onelake.dfs.fabric.microsoft.com/e5d2dce1-920c-4c8b-bb30-641e70bdd3d0/Files/rawAdventureWorks_Customers/AdventureWorks_Customers.csv")

display(df_cust)
```

```python
df_cust.printSchema()
```

```python
df_cust = df_cust.withColumn(
        "BirthDate",
        to_date("BirthDate", "M/d/yyyy")
    ).withColumn(
        "AnnualIncome",
        when(
            col("AnnualIncome").isNotNull(),
            regexp_replace(col("AnnualIncome"), "[$,]", "").cast("int")
        ).otherwise(None)
    )
```

```python
display(df_cust)
```

```python
df_cust = df_cust.withColumn("FullName",concat(col("Prefix"),lit(" "),col("FirstName"),lit(" "),col("LastName")))

display(df_cust)
```

```python
df_cust.write.format("delta")\
            .mode("append")\
            .option("path","Files/CUstomer")\
            .saveAsTable("Customer")
```

**Sales**

```python
df_sales2015= spark.read.format("csv")\
                   .option("header", True)\
                   .option("inferschema", True)\
                   .load("abfss://27c585a4-80a1-47b7-879c-9a8e709ed796@onelake.dfs.fabric.microsoft.com/e5d2dce1-920c-4c8b-bb30-641e70bdd3d0/Files/rawAdventureWorks_Sales_2015/AdventureWorks_Sales_2015.csv")

df_sales2016= spark.read.format("csv")\
                   .option("header", True)\
                   .option("inferschema", True)\
                   .load("abfss://27c585a4-80a1-47b7-879c-9a8e709ed796@onelake.dfs.fabric.microsoft.com/e5d2dce1-920c-4c8b-bb30-641e70bdd3d0/Files/rawAdventureWorks_Sales_2016/AdventureWorks_Sales_2016.cvsv")

df_sales2017= spark.read.format("csv")\
                   .option("header", True)\
                   .option("inferschema", True)\
                   .load("abfss://27c585a4-80a1-47b7-879c-9a8e709ed796@onelake.dfs.fabric.microsoft.com/e5d2dce1-920c-4c8b-bb30-641e70bdd3d0/Files/rawAdventureWorks_Sales_2017/AdventureWorks_Sales_2017.csv")
                   
```

```python
df_sales= df_sales2015.union(df_sales2016).union(df_sales2017)

display(df_sales)
```

```python
df_sales= df_sales.withColumn("OrderDate", to_date("OrderDate", "M/d/yyyy"))\
                  .withColumn("StockDate", to_date("StockDate", "M/d/yyyy"))
```

```python
display(df_sales)
```

```python
df_sales.write.format("delta")\
            .mode("append")\
            .option("path","Files/Sales")\
            .saveAsTable("Sales")
```
